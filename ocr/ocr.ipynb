```python
from pytesseract import Output
img2 = cv2.imread("D:/DL_project/ocr/dats/text.png")
```

```python
data = pytesseract.image_to_data(img2,output_type=Output.DICT)
```

```python
print(data.keys())
```

```python
print(data['text'])
```

```python
n_boxes=len(data['conf'])
n_boxes
```

```python
img2 = cv2.imread("D:/DL_project/ocr/dats/text.png")
for i in range(n_boxes):
    if int(data['conf'][i])>90:
        x,y,w,h = data['left'][i],data['top'][i],data['width'][i],data['height'][i]
        cv2.rectangle(img2,(x,y),(x+w,y+h),(0,255,0),1)

cv2.imshow('omage to data',img2)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img3 = cv2.imread("D:/DL_project/ocr/dats/invoice-sample.jpg")
```

```python
data1 = pytesseract.image_to_data(img3,output_type=Output.DICT)
```

```python
print(data1.keys())
```

```python
print(data1['text'])
```

```python
n_boxes1=len(data1['conf'])
n_boxes1
```

```python
date_pattern = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
email_pattern = '^[a-zA-Z0-9+_.-]+@[a-zA-Z0-9.-]+$'
```

```python
img3 = cv2.imread('D:/DL_project/ocr/dats/invoice-sample.jpg')
for i in range(n_boxes1):
    if int(data1['conf'][i])>60:
        if re.match(date_pattern,data1['text'][i]) or re.match(email_pattern,data1['text'][i]):

            x,y,w,h = data1['left'][i],data1['top'][i],data1['width'][i],data1['height'][i]
            cv2.rectangle(img3,(x,y),(x+w,y+h),(0,255,0),1)

        
cv2.imshow('image to data',img3)
cv2.waitKey()
cv2.destroyAllWindows()
```
